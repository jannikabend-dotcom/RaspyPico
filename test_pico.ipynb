```python
import matplotlib.pyplot as plt  # Plotten im Notebookimport subprocessimport sysdef run_on_pico(code: str, port: str = "auto", exec_timeout_s: float = 20.0):    cmd = [sys.executable, "-m", "mpremote", "connect", port, "exec", code]    proc = subprocess.run(cmd, capture_output=True, text=True, timeout=exec_timeout_s)    return proc.stdout.strip(), proc.stderr.strip()cmd = [sys.executable, "-m", "mpremote", "connect", "list"]ports = subprocess.run(cmd, capture_output=True, text=True, timeout=10)print(ports.stdout.strip() or "Keine Ports von mpremote erkannt.")if ports.stderr.strip():    print("stderr:\n", ports.stderr.strip())
```
